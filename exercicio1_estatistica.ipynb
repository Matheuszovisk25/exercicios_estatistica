# Exercicio 1

Uma pesquisa de opinião perguntou a 124 pessoas qual o antitérmico preferido entre as marcas A, B, C, D e E. Os resultado estão representados na tabela abaixo

image.png

Construa o gráfico de barras correspondente


```python
import pandas as pd
import seaborn as sns
```

```python
data = {"A": [45],
      "B": [32],
      "C": [23],
      "D": [15],
      "E": [9]}
```

```python
df = pd.DataFrame(data)
```

```python
import matplotlib.pyplot as plt
```

# Resposta

Este gráfico poderia ser destacado como um exemplo ruim no livro "Storytelling com Dados". mas pra essa resposta é o suficiente."

```python

plt.figure(figsize=(12,6))
plt.title("Resposta")
plt.ylim(0, 50)
plt.xlabel("Marcas")
plt.ylabel("N° pessoas")
plt.grid(True, linestyle='--', color='gray', linewidth=0.4)
sns.barplot(data=df)
```
